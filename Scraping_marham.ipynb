```python
import requests
from bs4 import BeautifulSoup

def extract_names_from_html(html):
    names = []
    soup = BeautifulSoup(html, "html.parser")
    doctor_cards = soup.find_all("h3", class_="text-underline")
    for card in doctor_cards:
        names.append(card.text.strip())
    return names

all_names = []
page = 1  # Move this outside the loop

while True:
    url = f"https://www.marham.pk/api/listing/load-more-doctors?page={page}&area=&availableToday=0&city=&discount=0&endTime=&feeRange=&gender=&isSpecilatyListing=1&lat=0&listingType=1&lng=0&mostBooked=0&nearMe=0&offset=0&onlineNow=0&physicalDiscount=&slug=&specialityId=52&startTime=&symptomID=0&topReviewed=0&variation=v2&videoConsultation=0"
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    data = response.json()

    current_names = []

    if page == 1:
        for key in ["html1", "html2"]:
            html_content = data.get(key, "")
            current_names += extract_names_from_html(html_content)
    else:
        html_content = data.get("html2", "")
        current_names = extract_names_from_html(html_content)

    # Add to final list
    all_names.extend(current_names)

    # Break if nothing new is found (meaning we've reached the end)
    if not current_names:
        break

    page += 1


# âœ… Print the results
for i, name in enumerate(all_names, start=1):
    print(i, name)
```

```python
import requests
from bs4 import BeautifulSoup

def get_total_pages(specialty_url):
    response = requests.get(specialty_url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Example: find all page number buttons
    page_links = soup.select("ul.pagination li a")

    # Extract numbers and get max
    page_numbers = []
    for link in page_links:
        try:
            num = int(link.text.strip())
            page_numbers.append(num)
        except:
            pass

    return max(page_numbers) if page_numbers else 1

# Example usage
url = "https://www.marham.pk/doctors/dermatologist"
print("Total pages:", get_total_pages(url))
```

```python
import requests
from bs4 import BeautifulSoup

def get_total_pages(specialty):
    base_url = f"https://www.marham.pk/doctors/{specialty}?page="
    page = 1

    while True:
        url = base_url + str(page)
        res = requests.get(url)
        soup = BeautifulSoup(res.content, 'html.parser')

        # Check if doctors/cards are found on this page
        doctor_cards = soup.select(".row.shadow-card")

        if not doctor_cards:
            break
        page += 1

    return page - 1  # Last valid page

# Example usage:
specialty = "dermatologist"
print("Total pages:", get_total_pages(specialty))
```
